```python
from eval import Evaluator
from scipy import ndimage
import numpy as np
```

```python
e = Evaluator()
r1, s1 = e.evaluate_traditional()
e.show(r1, s1, 'traditional')
r2, s2 = e.evaluate_psychometric()
e.show(r2, s2, 'psychometric')
r3, s3 = e.evaluate_wCon()
e.show(r3, s3, 'wCon')
r4, s4 = e.evaluate_wFre()
e.show(r4, s4, 'wFre')
r5, s5 = e.evaluate_cCon()
e.show(r5, s5, 'cCon')
r6, s6 = e.evaluate_cFre()
e.show(r6, s6, 'cFre')
r7, s7 = e.evaluate_sLen()
e.show(r7, s7, 'sLen')
r8, s8 = e.evaluate_wLen()
e.show(r8, s8, 'wLen')
r9, s9 = e.evaluate_oDen()
e.show(r9, s9, 'oDen')
print('***Variance***')
print('traditional : ', ndimage.variance(np.array(s1)))
print('psychometric: ', ndimage.variance(np.array(s2)))
print('wCon        : ', ndimage.variance(np.array(s3)))
print('wFre        : ', ndimage.variance(np.array(s4)))
print('cCon        : ', ndimage.variance(np.array(s5)))
print('cFre        : ', ndimage.variance(np.array(s6)))
print('sLen        : ', ndimage.variance(np.array(s7)))
print('wLen        : ', ndimage.variance(np.array(s8)))
print('oDen        : ', ndimage.variance(np.array(s9)))
```

```python
from scipy import stats

def f1list(scores):
    f = []
    for score in scores: 
        f.append(score[2])
    return f

pearson1 = stats.pearsonr(f1list(s1), f1list(s2))
pearson2 = stats.pearsonr(f1list(s3), f1list(s2))
pearson3 = stats.pearsonr(f1list(s4), f1list(s2))
pearson4 = stats.pearsonr(f1list(s5), f1list(s2))
pearson5 = stats.pearsonr(f1list(s6), f1list(s2))
pearson6 = stats.pearsonr(f1list(s7), f1list(s2))
pearson7 = stats.pearsonr(f1list(s8), f1list(s2))
pearson8 = stats.pearsonr(f1list(s9), f1list(s2))


print('***Correlation between committee judgement and attribute-aided difficulty***')
print('traditional : ', 'correlation: ', pearson1[0], 'p: ', pearson1[1])
print('wCon        : ', 'correlation: ', pearson2[0], 'p: ', pearson2[1])
print('cCon        : ', 'correlation: ', pearson3[0], 'p: ', pearson3[1])
print('wFre        : ', 'correlation: ', pearson4[0], 'p: ', pearson4[1])
print('cFre        : ', 'correlation: ', pearson5[0], 'p: ', pearson5[1])
print('sLen        : ', 'correlation: ', pearson6[0], 'p: ', pearson6[1])
print('wLen        : ', 'correlation: ', pearson7[0], 'p: ', pearson7[1])
print('oDen        : ', 'correlation: ', pearson8[0], 'p: ', pearson8[1])
```
